```python
import pandas as pd

wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None)
X, y = wine.iloc[:, 1:].values, wine[0].values
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
wine_pca = pca.fit_transform(wine)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter([p[0] for p in wine_pca], [p[1] for p in wine_pca], c = y)
plt.show()
```

```python
from sklearn.svm import SVC

clf = SVC()
clf.fit(wine_pca, y)
result = clf.predict(wine_pca)
```

```python
acc = 0
for i in range(len(y)):
    if y[i] == result[i]:
        acc += 1
acc /= len(y)       
print('Accuracy:', acc)
```
